# Module 1--- Compare RNA-seq at different time points (GEO accession: GSE206281)

## Myocardial infarction (MI) is one of the deadliest diseases in the world, and the changes at the molecular level after MI and the DNA methylation features are not clear. Understanding the molecular characteristics of the early stages of MI is of significance for the treatment of the disease. In this study, RNA-seq were performed on heart tissue from mouse models at multiple time points (0 h, 10 min,1 h, 6 h, 24 h and 72 h) to explore genetic and epigenetic features that influence MI progression.

## Read RNA-seq DNA obtained at different time points

```python
df1=pd.read_csv('./undergrad_chemistry_class_tutorial/GSE206282_RAW/GSM6249220_0h_FPKM.csv')
df2=pd.read_csv('./undergrad_chemistry_class_tutorial/GSE206282_RAW/GSM6249223_10m_FPKM.csv')
df3=pd.read_csv('./undergrad_chemistry_class_tutorial/GSE206282_RAW/GSM6249226_1h_FPKM.csv')
df4=pd.read_csv('./undergrad_chemistry_class_tutorial/GSE206282_RAW/GSM6249229_6h_FPKM.csv')
df5=pd.read_csv('./undergrad_chemistry_class_tutorial/GSE206282_RAW/GSM6249232_24h_FPKM.csv')
df6=pd.read_csv('./undergrad_chemistry_class_tutorial/GSE206282_RAW/GSM6249235_72h_FPKM.csv')
```

## For the convenience of data visualization, we only choose the first 100 genes in each dataset to compare gene expressions across different time points.

```python
df1=df1.iloc[0:100]
df2=df2.iloc[0:100]
df3=df3.iloc[0:100]
df4=df4.iloc[0:100]
df5=df5.iloc[0:100]
df6=df6.iloc[0:100]
```

```python
df2.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(1, figsize=(30, 55))
plt.subplot(611, title='0h')
plt.bar(df1.gene_symbol, df1.Mcon3)
plt.xticks(rotation=90)
plt.subplot(612, title='10min')
plt.bar(df2.gene_symbol, df2.M10m3)
plt.xticks(rotation=90)
plt.subplot(613, title='1h')
plt.bar(df3.gene_symbol, df3.Mlh3)
plt.xticks(rotation=90)
plt.subplot(614, title='6h')
plt.bar(df4.gene_symbol, df4.M6h3)
plt.xticks(rotation=90)
plt.subplot(615, title='24h')
plt.bar(df5.gene_symbol, df5.M24h3)
plt.xticks(rotation=90)
plt.subplot(616, title='72h')
plt.bar(df6.gene_symbol, df6.M72h3)
plt.xticks(rotation=90)
plt.show()
```

# Module 2--- Compare RNA-seq across different groups after drug treatment (GEO accession: GSE119088)

## The goals of this study are to compare NGS-derived transcriptome profiling (RNA-seq) to find out the difference between EZH2 inhibitor treatment and DMSO group in each of 3 cancer cell lines, and find the relationship between transcriptomes change and drug sensivitity.

## To load the transcriptome data for three cell lines (SMMC-7721, U2932, Pfeiffer):

```python
df7=pd.read_csv('./undergrad_chemistry_class_tutorial/S7721_transcriptome_change_07042022.csv')
df8=pd.read_csv('./undergrad_chemistry_class_tutorial/U2932_transcriptome_change_07042022.csv')
df9=pd.read_csv('./undergrad_chemistry_class_tutorial/Pfeif_transcriptome_change_07042022.csv')
```

```python
df8.head()
```

## For the convenience of data visualization, we still choose the first 100 genes and focus on their differential expressions.

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(1, figsize=(30, 30))
plt.subplot(311, title='SMMC-7721 Cell Line')
plt.bar(df7.Gene_symbol, df7.log2FoldChange)
plt.xticks(rotation=90)
plt.subplot(312, title='U2932 Cell Line')
plt.bar(df8.Gene_symbol, df8.log2FoldChange)
plt.xticks(rotation=90)
plt.subplot(313, title='Pfeiffer Cell Line')
plt.bar(df9.Gene_symbol, df9.log2FoldChange)
plt.xticks(rotation=90)
plt.show()
```
